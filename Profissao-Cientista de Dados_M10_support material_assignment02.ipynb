````python
#importando bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
#Lendo a base de dados
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# analise de renda com posse de imóveis
ax = sns.pointplot(data=df, x='posse_de_imovel', y='renda')
````

````python
#Correlação de renda entre possei de imoveis com educação
ax = sns.pointplot(data=df, x='educacao', y='renda', hue='posse_de_imovel')
````

````python
# correlação de maus pagadores com a posse de imoveis
ax = sns.pointplot(data=df, x='mau', y='renda', hue='posse_de_imovel')
````

nos dois gráficos temos um nível maior de renda ligado a posse de imóveis ou não, no ultimo gráfico percebemos que geralmente maus pagadores não possuem imoveis e recebem menos.

````python
#analise de renda com posse de veículo
ax = sns.pointplot(data=df, x='posse_de_veiculo', y='renda')
````

````python
ax = sns.pointplot(data=df, x='posse_de_veiculo', y='renda', hue='mau')
````

já aqui vemos maior discrepância entre os bons pagadores que possuem carros, o número de bons pagadores que possuem carro é maior que o número de bons pagadores q não possuem carro, então seria um otimo indicador, também percebemos o aumento significativo do salário de bons pagadores que possuem carro!
